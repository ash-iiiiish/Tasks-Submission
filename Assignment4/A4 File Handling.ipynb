# Task1

```python
# Task 1: Write sales records to a file

sales = [1200, 450, 980, 1500, 3000]

# Write mode
with open("sales_data.txt", "w") as file:
    for sale in sales:
        file.write(str(sale) + "\n")

# Reopen and print contents
with open("sales_data.txt", "r") as file:
    print("Task 1 Output:")
    print(file.read())
```

# Task2

```python
# Task 2: Read file in different ways

with open("sales_data.txt", "r") as file:
    # 1. Read entire file
    entire_content = file.read()
    print("Read using read():")
    print(entire_content)

with open("sales_data.txt", "r") as file:
    # 2. Read first line
    first_line = file.readline().strip()
    print("First line using readline():", first_line)

with open("sales_data.txt", "r") as file:
    # 3. Read all lines and convert to integers
    lines = file.readlines()
    sales_list = [int(line.strip()) for line in lines]

print("Sales list using readlines():", sales_list)
```

# Task3

```python
# Task 3: Append new sales

new_sales = [5000, 2500, 1700]

with open("sales_data.txt", "a") as file:
    for sale in new_sales:
        file.write(str(sale) + "\n")

# Reopen and print updated file
with open("sales_data.txt", "r") as file:
    updated_data = file.readlines()

print("Updated File Content:")
for line in updated_data:
    print(line.strip())

# Extra: total number of lines
print("Total number of lines:", len(updated_data))
```

# Task4

```python
# Task 4: Summary report

with open("sales_data.txt", "r") as file:
    sales = [int(line.strip()) for line in file.readlines()]

total_sales = sum(sales)
highest_sale = max(sales)
lowest_sale = min(sales)
average_sale = total_sales / len(sales)

print("\nSales Summary Report")
print("Total Sales:", total_sales)
print("Highest Sale:", highest_sale)
print("Lowest Sale:", lowest_sale)
print("Average Sale:", average_sale)
```

```python
# Task 6: Read File Safely

import os

filename = input("\nEnter filename to open: ")

if os.path.exists(filename):
    with open(filename, "r") as file:
        print("\nFile Contents:")
        print("-" * 20)
        print(file.read())
else:
    print("File not found. Please check the filename.")
```
