# Task3

```python
# 1. Create price dictionary with at least 6 products
price_dict = {
    "Laptop": 65000,
    "Phone": 30000,
    "Tablet": 25000,
    "Headphones": 3000,
    "Keyboard": 1500,
    "Mouse": 800
}

# 2a. Add a new product
price_dict["Smartwatch"] = 12000

# 2b. Update price of an existing product
price_dict["Phone"] = 28000

# 2c. Remove a product safely
product_to_remove = "Tablet"
if product_to_remove in price_dict:
    del price_dict[product_to_remove]
else:
    print(f"{product_to_remove} not found in price list")

# 3. Calculate average price
total_price = sum(price_dict.values())
average_price = total_price / len(price_dict)
print("Average price:", average_price)

# Extra (optional): Product with max and min price
max_product = max(price_dict, key=price_dict.get)
min_product = min(price_dict, key=price_dict.get)

print("Most expensive product:", max_product, price_dict[max_product])
print("Cheapest product:", min_product, price_dict[min_product])
```

# Task4

```python
# Given products list (example categories)
products = [
    ("Laptop", "Electronics"),
    ("Phone", "Electronics"),
    ("Keyboard", "Accessories"),
    ("Mouse", "Accessories"),
    ("Headphones", "Accessories"),
    ("Smartwatch", "Electronics")
]

# 1. Create catalog (product_name, price, category)
catalog = []
for product, category in products:
    if product in price_dict:
        catalog.append((product, price_dict[product], category))

print("Catalog:", catalog)

# 2. Create category_to_products dictionary
category_to_products = {}

for product, price, category in catalog:
    category_to_products.setdefault(category, []).append(product)

print("Category to Products:", category_to_products)

# 3. Print products from category with maximum products
max_category = max(category_to_products, key=lambda c: len(category_to_products[c]))

print("Category with maximum products:", max_category)
print("Products:", category_to_products[max_category])
```
