# Task1

```python
try:
    order_amount = int(input("Enter order amount: "))
except ValueError:
    print("Error: Please enter a numeric value.")
    exit()

if order_amount >= 2000:
    discount_rate = 0.15
elif order_amount >= 1500:
    discount_rate = 0.10
elif order_amount >= 1000:
    discount_rate = 0.07
else:
    discount_rate = 0

discount = order_amount * discount_rate
subtotal = order_amount - discount

tax = subtotal * 0.05   # 5% tax
final_amount = subtotal + tax

print("Order Amount:", order_amount)
print("Discount:", discount)
print("Subtotal:", subtotal)
print("Tax:", tax)
print("Final Amount:", final_amount)
```

# Task2

```python
orders = [1200, 2500, 800, 1750, 3000]
total_revenue = 0
discounted_orders = 0

print("Order | Discount% | Final Amount")
print("---------------------------------")

for order in orders:
    if order >= 2000:
        discount_rate = 0.15
    elif order >= 1500:
        discount_rate = 0.10
    elif order >= 1000:
        discount_rate = 0.07
    else:
        discount_rate = 0

    discount = order * discount_rate
    final_amount = order - discount

    if discount > 0:
        discounted_orders += 1

    total_revenue += final_amount

    print(order, " | ", int(discount_rate * 100), "% | ", final_amount)

print("---------------------------------")
print("Total Revenue after discounts:", total_revenue)
print("Orders with discount:", discounted_orders)
```

# Task3

```python
orders = []

while True:
    print("\nMenu:")
    print("1 - Add order amount")
    print("2 - Show all orders and total after discount")
    print("q - Quit")

    choice = input("Enter your choice: ")

    if choice == "1":
        amount = float(input("Enter order amount: "))
        orders.append(amount)

    elif choice == "2":
        if not orders:
            print("No orders yet.")
            continue

        total = sum(orders)
        discount = 0.1 * total   # 10% discount
        final_total = total - discount

        print("Orders:", orders)
        print("Total:", total)
        print("Discount:", discount)
        print("Final Total:", final_total)

    elif choice == "q":
        print("Exiting menu...")
        break

    else:
        print("Invalid choice. Try again.")
        continue
```

# Task5

```python
daily = [200, 150, 0, 400, 50, -1, 300]
total_sales = 0

for sale in daily:
    if sale == -1:
        print("Corrupted data found. Stopping processing.")
        break

    if sale == 0:
        print("No sales today. Skipping.")
        continue

    total_sales += sale
    print("Running total:", total_sales)

print("Final total sales:", total_sales)
```
