# Task1

```python
import pandas as pd

marks = pd.Series([78, 85, 90, 66, 72])

print("Series values:\n", marks.values)
print("Index:\n", marks.index)
print("Data type:", marks.dtype)

print("First element:", marks.iloc[0])
print("Last two elements:\n", marks.iloc[-2:])
```

# Task2

```python
print("Add 5:\n", marks + 5)
print("Subtract 2:\n", marks - 2)
print("Multiply by 1.05:\n", marks * 1.05)
print("Divide by 2:\n", marks / 2)
```

# Task3

```python
print("Maximum:", marks.max())
print("Minimum:", marks.min())
print("Sum:", marks.sum())
print("Mean:", marks.mean())

passed = marks.apply(lambda x: x >= 70)
print("Passed status:\n", passed)

print("Total students passed:", passed.sum())
```

# Task4

```python
students = {
    'Name': ['Amit', 'Neha', 'Rahul', 'Sneha', 'Pooja'],
    'Marks': [78, 85, 90, 66, 72],
    'Subject': ['Math', 'Math', 'Science', 'Science', 'Math']
}

df = pd.DataFrame(students)

print("First 3 rows:\n", df.head(3))
print("Last 2 rows:\n", df.tail(2))
print("Shape:", df.shape)
print("Columns:", df.columns)
```

# Task5

```python
print(df.info())
print(df.describe())
print(df.head())
print(df.tail())

df_sorted = df.sort_values(by='Marks', ascending=False)
df_sorted = df_sorted.reset_index(drop=True)

print("Sorted DataFrame:\n", df_sorted)
```

# Task6

```python
print("Marks > 75:\n", df[df['Marks'] > 75])
print("Math students:\n", df[df['Subject'] == 'Math'])

avg_marks = df['Marks'].mean()
print("Above average:\n", df[df['Marks'] > avg_marks])
print("Failed students:\n", df[df['Marks'] < 70])
```

# Task7

```python
print("Average marks per subject:\n", df.groupby('Subject')['Marks'].mean())
print("Count per subject:\n", df.groupby('Subject')['Marks'].count())
print("Max marks per subject:\n", df.groupby('Subject')['Marks'].max())
```

# Task8

```python
df.plot(x='Name', y='Marks', kind='bar')
df['Marks'].plot(kind='line')
df['Marks'].plot(kind='hist')
```

# Task9

```python
sales = {
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Revenue': [1200, 1500, 900, 2000, 1800]
}

sales_df = pd.DataFrame(sales)

print("Total revenue:", sales_df['Revenue'].sum())
print("Average revenue:", sales_df['Revenue'].mean())
print("Highest revenue day:\n", sales_df.loc[sales_df['Revenue'].idxmax()])
print("Above average days:\n", sales_df[sales_df['Revenue'] > sales_df['Revenue'].mean()])

sales_df.plot(x='Day', y='Revenue', kind='line')
```
