```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('C:/Users/kumar/OneDrive/Desktop/TRY 1/TuteDude-Course-1/Tasks-Submission/Assignment12/student_data.csv')
```

# Task1

```python
sns.relplot(
    data=df,
    x="total_bill",
    y="tip",
    hue="sex",
    kind="scatter"
)
plt.show()

# Same using line style
sns.relplot(
    data=df,
    x="total_bill",
    y="tip",
    hue="sex",
    kind="line"
)
plt.show()
```

# Task2

```python
# Line plot
sns.lineplot(data=df, x="total_bill", y="tip")
plt.show()

# Scatter style line plot
sns.relplot(data=df, x="total_bill", y="tip", kind="scatter")
plt.show()

# Faceted by category
sns.relplot(
    data=df,
    x="total_bill",
    y="tip",
    col="time",
    hue="sex",
    kind="scatter"
)
plt.show()
```

# Task3

```python
# Histogram
sns.histplot(df["total_bill"])
plt.show()

# KDE
sns.kdeplot(df["total_bill"])
plt.show()

# Rug
sns.rugplot(df["total_bill"])
plt.show()

# Histogram + KDE
sns.histplot(df["total_bill"], kde=True)
plt.show()
```

# Task4

```python
# Bivariate Histogram
sns.histplot(data=df, x="total_bill", y="tip")
plt.show()

# Bivariate KDE
sns.kdeplot(data=df, x="total_bill", y="tip", fill=True)
plt.show()
```

# Task5

```python
# Pair plot
sns.pairplot(df, hue="sex")
plt.show()

# Heatmap of correlation
corr = df.select_dtypes("number").corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

# Task6

```python
# Bar plot
sns.barplot(data=df, x="sex", y="total_bill")
plt.show()

# Box plot
sns.boxplot(data=df, x="sex", y="total_bill")
plt.show()

# Violin plot
sns.violinplot(data=df, x="sex", y="total_bill")
plt.show()

# Count plot
sns.countplot(data=df, x="day")
plt.show()
```

# Task7

```python
# Regression plot
sns.regplot(data=df, x="total_bill", y="tip")
plt.show()

# lmplot with hue
sns.lmplot(data=df, x="total_bill", y="tip", hue="sex")
plt.show()
```

# Task8

```python
# FacetGrid
g = sns.FacetGrid(df, col="time", hue="sex")
g.map(sns.scatterplot, "total_bill", "tip")
g.add_legend()
plt.show()

# Multi-plot dashboard
sns.relplot(data=df, x="total_bill", y="tip", hue="sex", col="time")
sns.catplot(data=df, x="day", y="total_bill", kind="bar")
sns.displot(data=df, x="total_bill", kde=True)
plt.show()
```
